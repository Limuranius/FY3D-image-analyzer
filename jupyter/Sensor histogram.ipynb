```python
from Deviations import Deviations
import matplotlib.pyplot as plt
plot_width = 15
plot_height = 30
plt.rcParams['figure.figsize'] = [plot_width, plot_height]
```

```python
fig, axes = plt.subplots(nrows=5, ncols=2)
year = 2023
channel = 14

for sensor in range(10):
    row = sensor // 2
    col = sensor % 2
    data = Deviations.get_dataframe(year=year, channel=channel, sensor=sensor)
    ax = axes[row, col]
    ax.title.set_text(f" {sensor}")
    data["deviation"].hist(bins=30, ax=ax)
```
